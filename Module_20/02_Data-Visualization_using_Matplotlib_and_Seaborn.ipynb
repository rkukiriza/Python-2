# Data Visualization
Data  visualization is the process of representing data graphically to reveal patterns, trends, correlations, and other outliers that might not be obvious in raw numbers. It is essential for data exploration, communication and decision making.

## Matplotlib
A versatile library for creating a wide range of plots, from simple lines to ncomplex 3D visuals. It gives you precise control over every element, colors, labels, sizes and more making it ideal for custom visualizations.

## Seaborn
A higher level library built on Matplotlib, designed for statistical data visualization. It simplifies creating complex plots, integrates easily with Pandas DataFrames and offers attractive default styles and color palettes.

Use case: use these tools to analyse data sets like sales records, scientific measurements e.g flower dimensions in this dataset or survey results to identify trends or relationships.

## Install/Import Matplotlib

```python
import matplotlib.pyplot as plt # import pyplot as plt
```

### Basic Set up

```python
set=['daniel','alfred','amah','divine','ruth','george']

scores=[80,90,60,70,85,74]
```

### Line Plot
We draw a line to show the number and how they change so to call a function we do:

```python
students=['daniel','alfred','amah','divine','ruth','george']

scores=[80,90,60,70,85,74]

plt.plot(students,scores,color='blue')
y_axis=scores
x_axis=students

# To make it visual(story telling functions)
# Title(picking function from module from which we plotted/picked)

plt.title('Student Test Score')
plt.xlabel('Students Name')
plt.ylabel('Scores')
plt.show() # text against visual
```

### Scatter Plot

```python
plt.scatter(students, scores, color='red')
y_axis=scores
x_axis=students
plt.title('Student Test Score')
plt.xlabel('Students Name')
plt.ylabel('Scores')
plt.show()
```

### Bar plot

```python
plt.bar(students, scores, color='green')
y_axis=scores
x_axis=students
plt.title('Student Test Score')
plt.xlabel('Student Name')
plt.ylabel('Scores')
plt.tight_layout
plt.show()
```

### Histogram Plot

Histograms show how many scores are in student range.

```python
plt.figure(figsize=(10,5)) # --> controls chart size
plt.hist(scores,bins=5, color='purple', edgecolor='black') # bins controls how score ranges are grouped.
y_axis=scores
x_axis=students
plt.title('Histogram of Student Scores') # 2 arguments cannot be passed in a histogram ie atudent names and scores, as student name
plt.xlabel('Number of students') # s are categorical and not used in histogram calculation.
plt.ylabel('score ranges')
plt.show()
```

# Seaborn

Is a super easy tool in Python that helps you draw pretty pictures called plots of your data. Its like a magic crayon box built on top of matplotlib, making graphs look nice and simple to make for beginners. 

Why use it? It runs numbers into cool, colorful pictures so you can see patterns fast.

## Install/Import Seaborn

```python
import seaborn as sns
import pandas as pd
```

### Basic set up

```python
students=['daniel','alfred','amah','divine','ruth','george']
scores=[80,90,60,70,85,74]

data=pd.DataFrame({'students':students,'scores':scores})
print(data)
```

### Line plot

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='students', y='scores', data=data, color='blue') # This draws a line chart
plt.title('Student Test Score')
plt.xlabel('Students Name')
plt.ylabel('Scores')
plt.show()
```

### Scatter Plot

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='students', y='scores', data=data, color='purple')
plt.title('Student Test Score')
plt.xlabel('Students Name')
plt.ylabel('Scores')
plt.show()
```

### Bar Plot

```python
plt.figure(figsize=(10,5))
sns.barplot(x='students', y='scores', data=data, color='red')
plt.title('Student Test Score')
plt.xlabel('Students Name')
plt.ylabel('Scores')
plt.show()
```

### Histogram Plot

 Shows how the scores are grouped together

```python
plt.figure(figsize=(10,5))
sns.histplot(data=data, x='scores', color='green')
x_axis=scores
plt.title('How Scores pile up')
plt.tight_layout
plt.show()
```
