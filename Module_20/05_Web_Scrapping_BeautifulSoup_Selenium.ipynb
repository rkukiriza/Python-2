# Web Scrapping with Beautiful Soup and Selenium

## What is Web Scrpping?
Web scrapping is like being a detective on the internet. Its a way to collect information (like text, prices or names) from websites by using a computer program. Instead of  coying and pasting by hand, you let Python grab the data for you.

## Beautiful Soup
Beautiful soup is a super simple Python tool that helpd you pull out data from Web pages. It reads the messy code od a website called HTML and makes it easy to find things like files, paragraphs, lists etc.

## Why  use it?
Saves time: Grabs lots of data fast like product prices or news headlines.

Fun for beginners: You can explore websites and collect data or cool information.

Useful: Get data for projectts like tracking toy prices or weather updates.

#### Scrapping data from Coin Desk

```python
# Set up Selenium WebDriver
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
# URL for coin Desk news
url="https://www.coindesk.com/"
driver.get(url)
# Wait for dynamic content to load
time.sleep(3) # Adjust based on page load time
# Parse page source with Beautiful Soup
soup=BeautifulSoup(driver.page_source, "html.parser")
# Find the news headlines
headlines=soup.find_all("h3", {"class": "card_title"})
for headline in headlines [:5]: # Limit to 5 headlines
    title=headline.text.strip()
    link=headline.find_parent("a") ["href"]
    print(f"headline:{title}\nLink:{url+link}\n")
# Close browser
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
```

```python
# Set up selenium with headles chrome for efficiency
options=webdriver.ChromeOptions()
options.add_argument("--headless")
options.add_argument("--disable-gpu")
options.add_argument("user_agent=Mozilla/5.0 (windowsNT 10.0;win64) chrome/91.0.4472.124")
driver=webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

try:
    url = "https://www.coindesk.com"
    driver.get(url)

    # Wait for headlines to load (adjust timeout if needed)
    try:
        WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.TAG_NAME, "h3"))
        )
    except Exception as e:
        print(f"Timeout waiting for headlines: {str(e)}")
        driver.quit()
        exit()

    # Parse page source with BeautifulSoup
    soup = BeautifulSoup(driver.page_source, "html.parser")

    # Find news headlines using a general fallback to h3 tags
    headlines = soup.find_all("h3")

    # Alternative: use a specific class selector if known (update class name)
    # headlines = soup.find_all("h3", {"class": "your_class_here"})

    print(f"Found {len(headlines)} headlines")

    if not headlines:
        print("No headlines found, check HTML structure or class name.")
        print("Sample HTML snippet:\n", soup.prettify()[:300])
        driver.quit()
        exit()

    # Extract and print up to 5 headlines
    for i, headline in enumerate(headlines[:5], 1):
        try:
            title = headline.text.strip()
            parent_link = headline.find_parent("a")
            link = parent_link["href"] if parent_link and parent_link.get("href") else "No link found"
            full_link = url + link if link.startswith("/") else link
            print(f"Headline {i}: {title}\nLink: {full_link}\n")
        except Exception as e:
            print(f"Error processing headline {i}: {str(e)}")

finally:
    # Clean up driver no matter what
    driver.quit()
```
